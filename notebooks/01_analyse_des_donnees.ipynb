### 2- Présentation des données

Le jeu de données utilisé est eudirectlapse, issu de la collection CAS Datasets. Il contient des informations relatives au renouvellement annuel de contrats d’assurance automobile pour une année et un assureur non spécifiés, et comprend 23 060 polices.

```python
#DATA_FILE = Path("..") / "data" / "raw" / "eudirectlapse.csv"
#df = pd.read_csv(DATA_FILE)

df = pd.read_csv('https://raw.githubusercontent.com/lucileb18/projet_SMOTE/refs/heads/main/data/raw/eudirectlapse.csv')
df.shape
```

La base de données contient 23 060 observations et 19 variables.

```python
df.head()
```

```python
df.info()
```

Dans ce jeu de données, on distingue une variable cible et un ensemble de variables explicatives. La variable lapse est une variable binaire indiquant la résiliation du contrat par le client.


Les variables explicatives décrivent les caractéristiques du souscripteur (âge, sexe, profession), du contrat (ancienneté, nombre de contrats, usage du véhicule), ainsi que des informations tarifaires (prime finale, prime précédente, prime technique, prime de marché). L’ensemble des variables explicatives comprend à la fois des variables numériques et des variables catégorielles.

Extrait de la documentation sur les variables :  
- `lapse` : variable binaire indiquant la résiliation du contrat.
  - Valeur 1 : le contrat est résilié par le client ;
  - Valeur 0 : le contrat est renouvelé.

- `polholder_age` : âge du souscripteur.
- `polholder_BMCevol` : évolution du coefficient bonus/malus (BMC) du souscripteur  
  (valeurs catégorielles : `"down"` lorsque le bonus diminue, `"stable"` lorsqu’il ne change pas, `"up"` lorsqu’il augmente).
- `polholder_diffdriver` : indicateur de différence entre le souscripteur et le conducteur.
- `polholder_gender` : genre du souscripteur.
- `polholder_job` : profession du souscripteur (`"medical"` ou `"normal"`).
- `policy_age` : ancienneté du contrat.
- `policy_caruse` : usage du véhicule assuré.
- `policy_nbcontract` : nombre de contrats détenus par le souscripteur auprès de l’assureur.
- `prem_final` : prime finale proposée au client pour le renouvellement.
- `prem_freqperyear` : fréquence de paiement de la prime par an.
- `prem_last` : prime payée lors de la dernière période de couverture.
- `prem_market` : proxy de la prime de marché.
- `prem_pure` : prime technique correspondant au risque théorique.
- `vehicl_age` : âge du véhicule.
- `vehicl_agepurchase` : âge du véhicule à l’achat.
- `vehicl_garage` : type de garage (variable catégorielle).
- `vehicl_powerkw` : puissance du véhicule en kilowatts (variable catégorielle).
- `vehicl_region` : région de résidence du souscripteur (exemple : Reg7, Reg4, ... ).

### 4 – Justification de l’utilisation de SMOTE


```python
TARGET_COL = "lapse"
```

```python
y = df[TARGET_COL]
print(y.value_counts())
print("\nProportions:")
print(y.value_counts(normalize=True))
```

La distribution de la variable d'intérêt lapse est déséquilibré. Environ 12,8 % des contrats correspondent à une résiliation contre 87,2 % de contrats non résiliés.

Dans un contexte assurantiel, où l’objectif est précisément d’identifier les contrats susceptibles de résilier, ce déséquilibre justifie l’étude de méthodes de rééquilibrage des classes telles que SMOTE.

```python
vc = y.value_counts()
plt.figure()
y.value_counts().plot(kind="bar")
plt.title("Distribution de la variable cible")
plt.xlabel("Classe")
plt.ylabel("Effectif")
plt.tight_layout()
plt.show()
```

### 5- Intégration de nouvelles variables

L'objectif sous-jacent de notre étude est d'améliorer la prédiction des résiliations de contrats d'assurance (lapse). On crée pour cela 3 nouvelles variables, susceptibles d'avoir une influence notable sur la décision de résiliation des assurés :
- variation de prime absolue
- variation de prime relative
- positionnement vs marché (prime / prime_market)

```python
# Variation de prime
df["prem_change_abs"] = df["prem_final"] - df["prem_last"]
df["prem_change_rel"] = ((df["prem_final"] / (df["prem_last"])) - 1)*100

# Écart au marché
df["prem_vs_market"] = df["prem_final"] / (df["prem_market"] )

df.replace([np.inf, -np.inf], np.nan, inplace=True)

print("Nouvelles variables ajoutées :", [
    "prem_change_abs", "prem_change_rel",
    "prem_vs_market"
])
```

```python
new_features = ["prem_change_abs", "prem_change_rel", "prem_vs_market"]
base_cols = ["prem_last", "prem_final", "prem_market"]  # utile pour diagnostiquer

# Comptage des NaN
nan_counts = df[new_features].isna().sum().sort_values(ascending=False)
print("NaN par colonne :")
print(nan_counts)

# Comptage des inf / -inf
inf_counts = df[new_features].apply(lambda s: np.isinf(s).sum()).sort_values(ascending=False)
print("\nInf/-Inf par colonne :")
print(inf_counts)
```

### 6- Analyse de la base de données

Nous souhaitons regarder la dépendance entre le pourcentage de résiliations et différents attributs au sein des contrats.

On commence par générer une matrice de corrélation entre les variables numériques pour observer la dépendance entre chaque variable et la résiliation, ainsi qu'entre chaque duo de variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure df is defined. If this cell is run independently, these lines might be needed:
# df = pd.read_csv('https://raw.githubusercontent.com/lucileb18/projet_SMOTE/refs/heads/main/data/raw/eudirectlapse.csv')

# Select only numerical columns for correlation matrix
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matrice de corrélation des variables numériques')
plt.show()
```

La matrice de corrélation apporte plusieurs informations importantes sur la dépendance de nos variables deux à deux :
- La probabilité de résiliation est faiblement corrélée avec toutes nos variables, c'est-à-dire d'un facteur 0.06 au maximum. Cependant, toutes les variables présentent une corrélation non-nulle et peuvent apporter une information pertinente pour le risque de résiliation.
- Les variables concernant une valeur de prime sont extrêmememnt corrélées, ce qui n'est pas aussi flagrant lorsqu'on observe les variables de variation de prime.

Étudions maintenant la dépendance des résiliation selon certaines variables.

```python
TARGET_COL = "lapse"
# Histogramme de la proportion de résiliation selon la prime
plt.figure(figsize=(12, 6))

sns.histplot(data=df, x='prem_final', hue=TARGET_COL, multiple='fill', stat='proportion', common_norm=True, kde=True, palette='viridis', legend=False)

plt.title('Résiliation selon la prime proposée au renouvellement')
plt.xlabel('Prime après renouvellement')
plt.ylabel('Proportion de résiliation')
plt.xticks(rotation=45, ha='right')
plt.xlim(100, 1600) # On exclut les valeurs extrêmes dans le graphique, car le peu d'occurences ne permet pas d'obtenir une proportion fiable
plt.ylim(0, 0.4)    # Pour augmenter la lisibilité de la courbe
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Trois paliers de proportion de résiliation se distinguent. D'une part, on observe que les assurés à primes faibles résilient peu leurs contrats. D'autre part, les assurées à prime moyenne présentent une proportion modérée. Ceux à prime élevée semblent enfin résilier plus souvent, bien que cette observation semble moins fiable.

L'existence de ces effets de seuil pourrait soulever l'intérêt de forêts aléatoires pour classer ces trois cas de figure, plutôt qu'une régression logistique.

D'autres relations entre plusieurs autres variables et la résiliation peuvent être tracées :

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='prem_change_abs', hue=TARGET_COL, multiple='fill', stat='proportion', common_norm=True, kde=False, palette='viridis', legend= False)

plt.title('Résiliation selon la variation absolue de prime')
plt.ylabel('Proportion de résiliation')
plt.xticks(rotation=45, ha='right')
plt.xlim(-15, 14)
plt.ylim(0, 0.35)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(2, 2, 2)

sns.histplot(data=df, x='prem_vs_market', hue=TARGET_COL, multiple='fill', stat='proportion', common_norm=True, kde=False, palette='viridis', legend= False)

plt.title('Résiliation selon la variation relative de prime')
plt.ylabel('Proportion de résiliation')
plt.xticks(rotation=45, ha='right')
plt.xlim(0.5, 1.75)
plt.ylim(0, 0.35)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(2, 2, 3)
sns.histplot(data=df, x='policy_age', hue=TARGET_COL, multiple='fill', stat='proportion', common_norm=True, kde=False, palette='viridis', bins=17,legend=False)

plt.title("Résiliation selon l'ancienneté du contrat")
plt.xlabel('Ancienneté du contrat')
plt.ylabel('Proportion de résiliation')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 0.35)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(2, 2, 4)
sns.histplot(data=df, x='polholder_age', hue=TARGET_COL, multiple='fill', stat='proportion', common_norm=True, kde=False, palette='viridis', bins=17, legend= False)

plt.title("Résiliation selon l'âge de l'assuré")
plt.xlabel('Age')
plt.ylabel('Proportion de résiliation')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 0.35)
plt.grid(axis='y', linestyle='--', alpha=0.7)
#plt.tight_layout()
plt.show()
```

Une tendance décroissante linéaire semble s'observer selon l'âge de l'assuré. Les personnes jeunes sont en effet enclines à comparer plus activement les différentes offres sur le marché que les personnes âgées.

Le premier graphique laisse penser que les profils qui parviennent à négocier la baisse de leur prime négocient également auprès d'autres assureurs ou sont convoités par ces derniers en tant que "bons risques", tandis que les assurés subissant une augmentation progressive de leur prime résilient relativement peu. En revanche, le graphique en haut à droite montre que si la prime diminue d'un facteur élevé (20% et plus), la résiliation est beaucoup plus faible car les nouvelles conditions deviennent très avantageuses.

Enfin, le graphique sur l'ancienneté du contrat montre également une dynamique de résiliation à plusieurs paliers, montrant un possible effet de seuil supplémentaire.

En conclusion, ces relations de dépendance non-linéaires montrées par cette étude seraient transposées plus facilement par la méthode des forêts aléatoires. Nous comparerons cette méthode avec celle de la régression logistique dans le dernier notebook.
